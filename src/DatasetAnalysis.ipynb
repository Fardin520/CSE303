## Balanced or imbalanced

### For firsst dataset(LoanData_Preprocessed_v1_1)

```python
df = pd.read_csv("D:\Codes\Python\MyJupyterPractice\CSE303\dataset\LoanData_Preprocessed_v1_1.csv")

# Count occurrences of each class in the target column
class_counts = df['default'].value_counts()
class_percentages = df['default'].value_counts(normalize=True) * 100
# Print the results
print("Class Distribution in 'default' column:")
print(class_counts)
print("\nPercentage Distribution:")
print(class_percentages)

# Check balance condition
threshold = 50  # Ideally, a balanced dataset has close to 50% in each class.
imbalance_ratio = max(class_percentages) / min(class_percentages)

if imbalance_ratio > 1.5:  # Arbitrary threshold, adjust based on requirement
    print("\nConclusion: The dataset is imbalanced".format(imbalance_ratio))
else:
    print("\nConclusion: The dataset is.")
```

# Calculating mean, median and variance:

### For first dataset()

```python
df = pd.read_csv("D:\Codes\Python\MyJupyterPractice\CSE303\dataset\LoanData_Preprocessed_v1_1.csv")
print("Analysis of Each Column:\n")

for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):  # Check if the column is numeric
        unique_vals = df[col].dropna().unique()  # Drop NaN values and get unique values
        
        # If the column has more than 10 unique values, consider it continuous
        if len(unique_vals) > 10:
            mean_val = df[col].mean()
            median_val = df[col].median()
            variance_val = df[col].var()
            std_dev_val = df[col].std()
            
            print(f"{col} (Continuous Numeric):")
            print(f" - Mean: {mean_val:.2f}")
            print(f" - Median: {median_val:.2f}")
            print(f" - Variance: {variance_val:.2f}")
            print(f" - Standard Deviation: {std_dev_val:.2f}")
        
        else:
            print(f"{col} (Discrete Numeric, treated as categorical):")
            value_counts = df[col].value_counts()
            percent_counts = df[col].value_counts(normalize=True) * 100
            
            print(" - Frequency counts:")
            print(value_counts.to_string())  # Show counts
            print(" - Percentage distribution:")
            print(percent_counts.to_string())  # Show percentages
    
    else:
        print(f"{col} (Categorical):")
        value_counts = df[col].value_counts()
        percent_counts = df[col].value_counts(normalize=True) * 100
        
        print(" - Frequency counts:")
        print(value_counts.to_string())
        print(" - Percentage distribution:")
        print(percent_counts.to_string())
    
    print("-" * 50)
```

# Ploting bar chart, line graph or pie chat where necessary. 

### For first dataset

```python
df = pd.read_csv("D:\Codes\Python\MyJupyterPractice\CSE303\dataset\LoanData_Preprocessed_v1_1.csv")

for col in df.columns:
    # Drop missing values for plotting to avoid issues
    clean_data = df[col].dropna()
    
    if pd.api.types.is_numeric_dtype(df[col]):
        # If many unique values, treat it as continuous and plot a line graph.
        if len(clean_data.unique()) > 10:
            plt.figure(figsize=(10, 5))
            plt.plot(clean_data, marker='o', linestyle='-', color='blue')
            plt.title(f"Line Graph for Continuous Column: {col}")
            plt.xlabel("Index")
            plt.ylabel(col)
            plt.grid(True)
            plt.show()
        else:
            # Otherwise, plot a bar chart for discrete numeric data.
            freq = clean_data.value_counts()
            plt.figure(figsize=(8, 5))
            freq.plot(kind="bar", color='skyblue')
            plt.title(f"Bar Chart for Discrete Column: {col}")
            plt.xlabel(col)
            plt.ylabel("Frequency")
            plt.show()
    else:
        # For non-numeric (categorical) columns, plot a pie chart.
        freq = clean_data.value_counts()
        plt.figure(figsize=(8, 8))
        freq.plot(kind="pie", autopct='%1.1f%%', title=f"Pie Chart for Categorical Column: {col}")
        plt.ylabel("")  # Remove y-label for clarity
        plt.show()
```

# Creating correlation heat-map

### For first dataset

```python
df = pd.read_csv("D:\Codes\Python\MyJupyterPractice\CSE303\dataset\LoanData_Preprocessed_v1_1.csv")

# Compute the correlation matrix for numeric columns
corr_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Create a heatmap with Seaborn
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Set title and display the plot
plt.title("Correlation Heatmap")
plt.show()
```
